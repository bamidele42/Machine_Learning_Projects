## Single Variable Linear regression Assignment
Canada's per capital income was given from year 1970 - 2016 use this information to predict per capital income for the year 2020.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df.tail()
```

```python
# Lets plot the points
plt.xlabel("year")
plt.ylabel("per capital income (US $)")
plt.scatter(df["year"], df["per capita income (US$)"])
```

```python
# Instantiating a linear regression object
reg = linear_model.LinearRegression()
reg.fit(df[["year"]], df["per capita income (US$)"])
```

```python
reg.predict(np.array(2020).reshape(-1, 1))
```

```python
# Checking the coefficient of the line
reg.coef_
```

```python
# Checking the intercept
reg.intercept_
```

```python
%matplotlib inline
plt.xlabel("year", fontsize= 20)
plt.ylabel("per capita income", fontsize= 20)

plt.plot(df["year"], reg.predict(df[["per capita income (US$)"]]), color="blue")
#plt.scatter(df["year"], df["per capita income (US$)"], color="red")
```

The regression line ( There is always a slope and intercept represented by m and b)
- year = m * per capita income + b
